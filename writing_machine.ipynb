```python
import random
```

```python
import pandas as pd
import numpy as np
```

```python
words = open('words.txt','r')
words_list = words.readlines()
```

```python
def random_word(l):
    random_index = random.randint(0,len(l)-1)
    return l[random_index].replace("\n", "")
random_word(words_list)
```

```python
random_list=[]
for i in range (0,256): 
    random_list.append(random_word(words_list))
random_list
```

```python
num = np.array(random_list)
```

```python
reshaped = num.reshape(16,16)
```

```python
df = pd.DataFrame(reshaped)
df
```

```python
my_sentence = []
```

```python
print('To write a verse select a row from 0 - 15: ')
row=int(input())
print('Select a word on that row by column 0 - 15: ')
column=int(input())
try:
    my_sentence.append(df.loc[row,column-1])
except:
    pass
my_sentence.append(df.loc[row,column])
try:
    my_sentence.append(df.loc[row,column+1])
except:
    pass
verse=''
for word in my_sentence:
    verse+=word+' '
print(verse)
```
